```python
from pyproj import CRS
src = CRS.from_proj4("+proj=utm +zone=47 +datum=WGS84 +units=m +no_defs")
dest = CRS.from_epsg(4326)
```

```python
from pyproj import Transformer
transformer = Transformer.from_crs(src, dest, always_xy=True)
lng, lat = transformer.transform(325210.246070927, 620860.599734336)
```

```python
import json
fin = open("Province/TH_Province.geojson", encoding="utf-8", errors='ignore')
geo = json.load(fin)
fin.close()
```

```python
fin = open("Province_map_thai_eng.json", encoding="utf-8-sig")
provinces = json.load(fin)
fin.close()
```

```python
provNameTh = {}
for prov in provinces:
    k = prov["en"].upper().replace(" ", "")
    provNameTh[k] = prov["th"]
```

```python
from tqdm import tqdm

debug = 20
for feat in tqdm(geo["features"], total=len(geo["features"])):
#     print(feat)
    name = feat["properties"]["PROV_NAME"].replace(" ", "")
    assert(name in provNameTh)
        
    feat["properties"]["PROV_NAMT"] = provNameTh[name]
    
    newcoords = []
    for coord in feat["geometry"]["coordinates"]:
        newcoord = []
        for c1 in coord:
            if len(c1) !=2:
                newc1 = []
                for c2 in c1:
                    assert(len(c2)==2)
                    newc2 = transformer.transform(c2[0], c2[1])
                    
                    if debug > 0:
                        print(c2, newc2)
                        debug -= 1
                    newc1.append(newc2)
                newcoord.append(newc1)
            else:
                newc1 = transformer.transform(c1[0], c1[1])
                newcoord.append(newc1)
                
                if debug > 0:
                    print(c1, newc1)
                    debug -= 1
                    
            assert(len(newc1)==len(c1))
        assert(len(newcoord)==len(coord))
        newcoords.append(newcoord)
    feat["geometry"]["coordinates"] = newcoords
"DONE"
```

```python
fout = open("TH_Province.geojson", "w", encoding="utf-8")
json.dump(geo, fout)
fout.close()
```
